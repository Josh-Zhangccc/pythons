# 遍历算法 Traversal
## 数组的遍历

```python
array = [0,3,4,6,8,2]
# 索引遍历(也可以直接遍历元素)
for i in range(len(array)):
    print(array[i])

# 枚举遍历
for index,value in enumerate(array):
    print("index:",index,"value:",value)

# 反向遍历
for i in range(len(array)-1,-1,-1):
    print(array[i])

# 步长遍历
for i in range(0,len(array),2):
    print(array[i])
```

## 链表

```python
class ListNode:
    """链表节点"""
    def __init__(self, value):
        self.value = value
        self.next = None

def create_linked_list(values):
    """创建链表"""
    if not values:
        return None
    
    head = ListNode(values[0])
    current = head
    
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    
    return head

def linked_list_traversal_demo():
    """链表遍历的各种方式"""
    # 创建链表: 1 -> 2 -> 3 -> 4 -> 5
    head = create_linked_list([1, 2, 3, 4, 5])
    
    print("=== 链表遍历 ===")
    
    # 方式1：基本遍历
    print("1. 基本遍历:")
    current = head
    position = 0
    while current:
        print(f"  位置 {position}: 值 {current.value}")
        current = current.next
        position += 1
    
    # 方式2：递归遍历
    print("\n2. 递归遍历:")
    def recursive_traverse(node, depth=0):
        if node is None:
            return
        indent = "  " * depth
        print(f"{indent}深度 {depth}: 值 {node.value}")
        recursive_traverse(node.next, depth + 1)
    
    recursive_traverse(head)
    
    # 方式3：带操作的遍历（比如求和）
    print("\n3. 带操作的遍历 - 求和:")
    current = head
    total = 0
    while current:
        total += current.value
        print(f"  当前值: {current.value}, 累计和: {total}")
        current = current.next
    
    print(f"  最终和: {total}")

linked_list_traversal_demo()
```
