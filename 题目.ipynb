## 递归
### 斐波那契数列第n项(Fibonacci Sequence)
题目描述： 编写一个递归函数 fibonacci(n)，返回斐波那契数列的第n项。斐波那契数列定义如下：
>F(0) = 0
>
>F(1) = 1
>
>F(n) = F(n-1) + F(n-2) (当 n ≥ 2)

```python
def F(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return F(n-1)+F(n-2)
print(F(n=int(input('请输入参数n'))))
```
